```python
rom pathlib import Path
import pandas as pd

# === Pfade anpassen ===
BASE_DIR = Path.home() / "Desktop" / "Masterarbeit" / "Code" / "data"
OUT_DIR  = BASE_DIR / "merged_groups"
OUT_DIR.mkdir(parents=True, exist_ok=True)

# === Gruppen & Model Codes ===
GROUPS = {
    "Nike_SM": ["HJ9198", "FD2722", "DH2987", "DJ6258", "DN3577"],
    "Nike_SF": ["DD1873", "DH3158", "FB2208", "FZ1689", "HM6804"],
    "Nike_CM": ["HQ2592", "HV8154", "DR7882", "FD1146", "DD1391"],
    "Nike_CF": ["DM0113", "IM1652", "DV1238", "HQ7540", "HQ2593"],
}

def read_csv_robust(path: Path) -> pd.DataFrame:
    # robust gegen Encoding-Probleme
    for enc in ["utf-8", "utf-8-sig", "cp1252", "latin-1"]:
        try:
            return pd.read_csv(path, encoding=enc, engine="python")
        except Exception:
            pass
    # wenn alles fehlschlägt, einmal "normal" werfen lassen (zeigt den echten Fehler)
    return pd.read_csv(path, engine="python")

def merge_group(group_name: str, codes: list[str]) -> pd.DataFrame:
    frames = []
    missing = []

    for code in codes:
        file_path = BASE_DIR / f"nike_reviews_{code}_EN.csv"
        if not file_path.exists():
            missing.append(str(file_path))
            continue

        df = read_csv_robust(file_path)

        # Metadaten, damit du später sauber filtern kannst
        df["model_code"] = code
        df["nike_group"] = group_name
        df["brand"] = "Nike"
        df["sustainability"] = "Sustainable" if group_name.startswith("Nike_S") else "Conventional"
        df["gender"] = "Male" if group_name.endswith("M") else "Female"

        frames.append(df)

    if missing:
        print(f"\n⚠️ Fehlende Dateien in {group_name}:")
        for m in missing:
            print(" -", m)

    if not frames:
        raise ValueError(f"Keine Dateien für {group_name} eingelesen (alle fehlen oder Fehler beim Lesen).")

    merged = pd.concat(frames, ignore_index=True, sort=False)

    # Optional: Duplikate entfernen, falls IDs mehrfach vorkommen
    if "id" in merged.columns:
        merged = merged.drop_duplicates(subset=["id"], keep="first")

    out_path = OUT_DIR / f"{group_name}_Total.csv"
    merged.to_csv(out_path, index=False)

    print(f"✅ {group_name}: gespeichert -> {out_path.name} | Rows: {len(merged)} | Cols: {merged.shape[1]}")
    return merged

# === Nur die 4 Gruppen bauen ===
group_dfs = {}
for group_name, codes in GROUPS.items():
    group_dfs[group_name] = merge_group(group_name, codes)

print("\nFertig. Erzeugte Dateien:")
for g in GROUPS:
    print(" -", (OUT_DIR / f"{g}_Total.csv").as_posix())
```

```python
from pathlib import Path
import pandas as pd

# === Ordner, in dem Nike_Total.csv & Adidas_Total.csv liegen ===
DATA_DIR = Path.home() / "Desktop" / "Masterarbeit" / "Code" / "Data"

NIKE_PATH  = DATA_DIR / "Nike_Total.csv"
ADIDAS_PATH = DATA_DIR / "Adidas_Total.csv"

nike = pd.read_csv(NIKE_PATH)
adidas = pd.read_csv(ADIDAS_PATH)

def pick_col(df: pd.DataFrame, candidates: list[str]) -> str | None:
    """Find first matching column (case-insensitive)."""
    m = {c.lower(): c for c in df.columns}
    for cand in candidates:
        if cand.lower() in m:
            return m[cand.lower()]
    return None

def to_bool_series(s: pd.Series) -> pd.Series:
    """Convert typical string/0-1/bool representations to pandas boolean dtype."""
    if s is None:
        return None
    if pd.api.types.is_bool_dtype(s):
        return s.astype("boolean")
    if pd.api.types.is_numeric_dtype(s):
        return s.map(lambda x: True if x == 1 else (False if x == 0 else pd.NA)).astype("boolean")

    x = s.astype("string").str.strip().str.lower()
    mapping = {
        "true": True, "1": True, "yes": True, "y": True, "t": True,
        "false": False, "0": False, "no": False, "n": False, "f": False,
    }
    return x.map(mapping).astype("boolean")

def parse_mixed_dates(s: pd.Series) -> pd.Series:
    """
    Handles your two known formats:
    - Nike: 20.01.2026  (DD.MM.YYYY)
    - Adidas: 2026-02-06T09:16:50.000+00:00  (ISO timestamp, often with timezone)
    Returns timezone-aware datetimes in UTC when possible.
    """
    x = s.astype("string").str.strip()

    # Nike: DD.MM.YYYY
    is_nike = x.str.match(r"^\d{1,2}\.\d{1,2}\.\d{4}$")
    dt = pd.to_datetime(x.where(is_nike), format="%d.%m.%Y", errors="coerce")

    # Adidas (and any remaining): ISO (parse as UTC if TZ present)
    dt = dt.fillna(pd.to_datetime(x.where(~is_nike), errors="coerce", utc=True))

    return dt

def standardize(df: pd.DataFrame, brand: str) -> pd.DataFrame:
    df = df.copy()

    id_col = pick_col(df, ["id", "review_id"])
    date_col = pick_col(df, ["date", "created_at", "created", "timestamp"])
    rating_col = pick_col(df, ["rating", "stars", "star_rating", "score"])
    title_col = pick_col(df, ["title", "headline"])
    text_col = pick_col(df, ["text", "review", "body", "content"])
    locale_col = pick_col(df, ["locale", "language", "lang"])
    inc_col = pick_col(df, ["incentivized", "isIncentivized", "incentivized_review"])
    sust_col = pick_col(df, ["sustainability", "Sustainability"])
    gender_col = pick_col(df, ["gender", "Gender"])

    # parse date robustly (no warning)
    if date_col:
        dt = parse_mixed_dates(df[date_col])
        # convert to plain date string YYYY-MM-DD (timezone-safe)
        date_out = dt.dt.strftime("%Y-%m-%d")
    else:
        date_out = pd.Series([pd.NA] * len(df), dtype="string")

    rating_out = pd.to_numeric(df[rating_col], errors="coerce") if rating_col else pd.Series([pd.NA]*len(df))
    inc_out = to_bool_series(df[inc_col]) if inc_col else pd.Series([pd.NA]*len(df), dtype="boolean")

    out = pd.DataFrame({
        "Brand": brand,
        "ID": df[id_col].astype("string") if id_col else pd.Series([pd.NA]*len(df), dtype="string"),
        "Date": date_out.astype("string"),
        "Rating": rating_out,
        "Title": df[title_col].astype("string") if title_col else pd.Series([pd.NA]*len(df), dtype="string"),
        "Text": df[text_col].astype("string") if text_col else pd.Series([pd.NA]*len(df), dtype="string"),
        "Locale": df[locale_col].astype("string") if locale_col else pd.Series([pd.NA]*len(df), dtype="string"),
        "Incentivized": inc_out,
        "Sustainability": df[sust_col].astype("string") if sust_col else pd.Series([pd.NA]*len(df), dtype="string"),
        "Gender": df[gender_col].astype("string") if gender_col else pd.Series([pd.NA]*len(df), dtype="string"),
    })

    return out

nike_clean = standardize(nike, "Nike")
adidas_clean = standardize(adidas, "Adidas")

all_reviews = pd.concat([nike_clean, adidas_clean], ignore_index=True, sort=False)

# Duplikate entfernen (sicher: Brand+ID)
all_reviews = all_reviews.drop_duplicates(subset=["Brand", "ID"], keep="first")

# Spaltenreihenfolge fix
cols = ["Brand", "ID", "Date", "Rating", "Title", "Text", "Locale", "Incentivized", "Sustainability", "Gender"]
all_reviews = all_reviews[cols]

OUT_PATH = DATA_DIR / "All_Reviews.csv"
all_reviews.to_csv(OUT_PATH, index=False)

print("✅ Saved:", OUT_PATH)
print("Rows/Cols:", all_reviews.shape)
print("\nVerteilung nach Brand/Sustainability/Gender:")
print(all_reviews.groupby(["Brand", "Sustainability", "Gender"]).size())
```
