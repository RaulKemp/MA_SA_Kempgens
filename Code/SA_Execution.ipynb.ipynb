```python
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score

df_siebert = pd.read_csv(r"C:\Users\Raul\Desktop\MA\reviews_full_siebert_results.csv")
df_vader = pd.read_csv(r"C:\Users\Raul\Desktop\MA\reviews_vader_results.csv")

report_siebert = classification_report(df_siebert['Sentiment'], df_siebert['Predicted_Sentiment'], output_dict=True)
report_vader = classification_report(df_vader['Sentiment'], df_vader['VADER_Sentiment'], output_dict=True)

metrics = {
    'Method': ['SieBERT', 'VADER'],
    'Accuracy': [
        round(accuracy_score(df_siebert['Sentiment'], df_siebert['Predicted_Sentiment']), 4),
        round(accuracy_score(df_vader['Sentiment'], df_vader['VADER_Sentiment']), 4)
    ],
    'Precision (Negative)': [
        round(report_siebert['negative']['precision'], 4),
        round(report_vader['negative']['precision'], 4)
    ],
    'Recall (Negative)': [
        round(report_siebert['negative']['recall'], 4),
        round(report_vader['negative']['recall'], 4)
    ],
    'F1-Score (Negative)': [
        round(report_siebert['negative']['f1-score'], 4),
        round(report_vader['negative']['f1-score'], 4)
    ],
    'Precision (Positive)': [
        round(report_siebert['positive']['precision'], 4),
        round(report_vader['positive']['precision'], 4)
    ],
    'Recall (Positive)': [
        round(report_siebert['positive']['recall'], 4),
        round(report_vader['positive']['recall'], 4)
    ],
    'F1-Score (Positive)': [
        round(report_siebert['positive']['f1-score'], 4),
        round(report_vader['positive']['f1-score'], 4)
    ],
    'Macro F1': [
        round(report_siebert['macro avg']['f1-score'], 4),
        round(report_vader['macro avg']['f1-score'], 4)
    ]
}

df_table = pd.DataFrame(metrics).set_index('Method')
print(df_table.to_string())

df_table.to_csv(r"C:\Users\Raul\Desktop\MA\evaluation_metrics_table.csv")
print("\nTabelle gespeichert!")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

df_siebert = pd.read_csv(r"C:\Users\Raul\Desktop\MA\reviews_full_siebert_results.csv")
df_original = pd.read_csv(r"C:\Users\Raul\Desktop\MA\All_Reviews_Organic.csv")
df = df_siebert.merge(df_original[['ID', 'Sustainability', 'Gender']], on='ID', how='left')
n
def chi_square_test(df, group_col, group1, group2, label=""):
    g1 = df[df[group_col] == group1]['Predicted_Sentiment'].value_counts()
    g2 = df[df[group_col] == group2]['Predicted_Sentiment'].value_counts()
    contingency_table = pd.DataFrame({group1: g1, group2: g2}).fillna(0)
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    n = contingency_table.values.sum()
    cramers_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))
    print(f"\n=== {label} ===")
    print(f"Contingency Table:\n{contingency_table}")
    print(f"Chi2: {chi2:.4f} | p-value: {p:.4f} | df: {dof}")
    print(f"Cramér's V: {cramers_v:.4f}")
    if p < 0.05:
        print("→ Signifikant (p < 0.05)")
    else:
        print("→ Nicht signifikant (p >= 0.05)")

df_sustainable = df[df['Sustainability'] == 'Sustainable']

chi_square_test(df, 'Brand', 'Nike', 'Adidas', label="Brand Overall: Nike vs. Adidas")
chi_square_test(df_sustainable, 'Brand', 'Nike', 'Adidas', label="H4: Nike Sustainable vs. Adidas Sustainable")
chi_square_test(df, 'Sustainability', 'Sustainable', 'Conventional', label="H2: Sustainable vs. Conventional")
chi_square_test(df_sustainable, 'Gender', 'Female', 'Male', label="H3: Female vs. Male (Sustainable only)")
```

```python
# VADER Analyse

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report
import pandas as pd
import time

df_final = pd.read_csv(r"C:\Users\Raul\Desktop\MA\All_Reviews_preprocessed.csv")

analyzer = SentimentIntensityAnalyzer()

start = time.time()

vader_results = []
for text in df_final['Combined_Text']:
    score = analyzer.polarity_scores(text)
    vader_results.append('positive' if score['compound'] >= 0.05 else 'negative')

elapsed = time.time() - start

df_final['VADER_Sentiment'] = vader_results
df_final['VADER_Correct'] = df_final['Sentiment'] == df_final['VADER_Sentiment']

print(f"VADER Accuracy: {df_final['VADER_Correct'].mean():.2%}")
print(f"VADER Laufzeit: {elapsed:.1f} Sekunden ({elapsed/60:.1f} Minuten)")

df_final.to_csv(r"C:\Users\Raul\Desktop\MA\All_Reviews_vader_results.csv", index=False)
print("VADER Ergebnisse gespeichert!")

df_siebert = pd.read_csv(r"C:\Users\Raul\Desktop\MA\All_Reviews_siebert_results.csv")

print("\n=== VADER ===")
print(classification_report(df_final['Sentiment'], df_final['VADER_Sentiment']))

print("=== SieBERT ===")
print(classification_report(df_siebert['Sentiment'], df_siebert['Predicted_Sentiment']))

df_combined = df_siebert[['Brand', 'ID', 'Rating', 'Sentiment', 'Combined_Text',
                           'Predicted_Sentiment', 'Confidence', 'Correct']].copy()
df_combined.rename(columns={
    'Predicted_Sentiment': 'SieBERT_Sentiment',
    'Confidence': 'SieBERT_Confidence',
    'Correct': 'SieBERT_Correct'
}, inplace=True)

df_combined['VADER_Sentiment'] = df_final['VADER_Sentiment'].values
df_combined['VADER_Correct'] = df_final['VADER_Correct'].values

df_combined.to_csv(r"C:\Users\Raul\Desktop\MA\All_Reviews_combined_results.csv", index=False)
print("\nCombined Datei gespeichert!")
```
